Customizing Plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a plot with custom styles
plt.plot(x, y, color='red', linestyle='--', linewidth=2, marker='o', markersize=6)

# Customize the plot
plt.xlabel('x', fontsize=14)
plt.ylabel('sin(x)', fontsize=14)
plt.title('Customized Sine Wave Plot', fontsize=16, fontweight='bold')
plt.grid(True)

plt.show()
```

Subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x1 = np.linspace(0, 10, 100)
y1 = np.sin(x1)
x2 = np.linspace(0, 5, 50)
y2 = np.cos(x2)

# Create a figure with two subplots
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# Plot the first subplot
axs[0].plot(x1, y1, color='blue')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')
axs[0].set_title('Sine Wave')

# Plot the second subplot
axs[1].plot(x2, y2, color='red')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')
axs[1].set_title('Cosine Wave')

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Create 3D data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a plot with custom styles
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='darkblue', linewidth=2, linestyle='-', label='Sine Wave')

# Customize axes and title
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.title('Customizing a Sine Wave Plot', fontsize=16)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure with two subplots
fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(8, 6))

# Plot the first subplot
axs[0].plot(x, y1, color='blue', label='Sine')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Sine Wave')
axs[0].legend()

# Plot the second subplot
axs[1].plot(x, y2, color='red', label='Cosine')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')
axs[1].set_title('Cosine Wave')
axs[1].legend()

plt.tight_layout()
plt.show()
```
